# **Data Preprocessing & Cleaning**

# Load the Heart Disease UCI dataset into a Pandas DataFrame

```python
df = pd.concat([features,target] ,axis= 1)
df
```

```python
df.info()
```

#  Data Validation

```python
for col in df.columns:
  print(f"range of {col}  : {df[col].min()} ===> {df[col].max()} ...nunique = {df[col].nunique()}")
```

# Handle missing values (imputation or removal).

```python
df.columns
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
df['ca'].fillna(df['ca'].mode()[0], inplace=True)
df.loc[(df['thal'].isnull()) & (df['ca'] == 0), 'thal'] = 3
df['thal'].fillna(df['thal'].mode()[0], inplace=True)
```

```python
df.isna().sum().sum()
```

```python
df["num"] = df["num"].apply(lambda value: 1 if value > 0 else 0)
```

```python
df["num"].nunique()
```

```python
categorical_cols = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal']
```

```python
df_encoded = pd.get_dummies(df, columns=categorical_cols, drop_first=True, dtype=float)
```

```python
df_encoded
```

```python
from sklearn.preprocessing import RobustScaler

numerical_cols = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']

scaler = RobustScaler()

df_encoded[numerical_cols] = scaler.fit_transform(df_encoded[numerical_cols])

```
